```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

Base_Dados = pd.read_csv('/Experience-Salary.csv')

col_experiencia = "exp(in months)"
col_salario = "salary(in thousands)"
Base_Dados.rename(columns={col_experiencia: "Experiencia", col_salario: "Salario"}, inplace=True)

Base_Dados["Experiencia"] = Base_Dados["Experiencia"] / 12

X = Base_Dados["Experiencia"].values.reshape(-1, 1)
y = Base_Dados["Salario"].values

X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size=0.2, random_state=0)

modelo = LinearRegression()
modelo.fit(X_treino, y_treino)

previsoes = modelo.predict(X_teste)

mse = mean_squared_error(y_teste, previsoes)
print(f'Erro quadrático médio: {mse:.2f}')

plt.scatter(X_teste, y_teste, color='blue', label='Dados reais')
plt.plot(X_teste, previsoes, color='red', linewidth=2, label='Previsões')
plt.xlabel('Anos de Experiência')
plt.ylabel('Salário (milhares)')
plt.legend()
plt.title('Regressão Linear: Experiência vs. Salário')
plt.show()
```
